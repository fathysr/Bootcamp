Uzun Soluklu Filmlerin Oluştuğu Diller


```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

df = pd.read_csv('C:/Users/Fatih/Desktop/bootcamp_mysia/NetflixOriginals.csv' , encoding="ISO-8859-1")
rl = df.groupby("Language").agg({"Runtime": "mean"}).sort_values(by="Runtime", ascending=False)
print(rl)


Runtime=df['Runtime']
Language=df['Language']
x=list(Runtime)
y=list(Language)
fig, ax = plt.subplots(figsize=(12, 6))
plt.bar(x, y, color = 'y', width = 0.4, label = "Lng")
plt.xlabel('Runtime', color ='r')
plt.ylabel('Language', color ='r')
plt.title('Uzun Soluklu Filmler', color = 'r')
plt.legend()
plt.show()
```
